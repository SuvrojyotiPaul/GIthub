## Q3

## Find the average gpa of all students.

```python
c.execute('''SELECT AVG(students.gpa) 
FROM courses
INNER JOIN students 
ON students.student_id = courses.student_id;''')

df3=pd.DataFrame(c.fetchall(),columns=['Average_Gpa'])
df3
```

## SELECT the student with the maximum gpa, display only their student_id, major and gpa

```python
c.execute('''SELECT students.student_id, students.major,MAX(students.gpa)
FROM courses
INNER JOIN students 
ON students.student_id = courses.student_id
WHERE students.gpa=(SELECT MAX(gpa) FROM students);''')

df3=pd.DataFrame(c.fetchall(),columns=['Student_id','Major','Gpa'])
df3
```

## SELECT the student with the minimum gpa, display only their student_id, major and gpa

```python
c.execute('''SELECT students.student_id, students.major,MIN(students.gpa)
FROM courses
INNER JOIN students 
ON students.student_id = courses.student_id
WHERE students.gpa=(SELECT MIN(gpa) FROM students);''')

df3=pd.DataFrame(c.fetchall(),columns=['Student_id','Major','Gpa'])
df3
```

## SELECT the students with a gpa greater than 3.6 in the majors of "Physics" and "Engineering", display only their student_id, major and gpa

```python
c.execute('''SELECT students.student_id, students.major,students.gpa
FROM courses
INNER JOIN students 
ON students.student_id = courses.student_id
WHERE students.gpa>3.6 AND students.major IN ('Physics','Engineering');''')

df3=pd.DataFrame(c.fetchall(),columns=['Student_id','Major','Gpa'])
df3
```

## Group the students by their major and retrieve the average grade of each major.

```python
c.execute(''' SELECT students.major, AVG(students.gpa)
FROM courses
INNER JOIN students 
ON students.student_id = courses.student_id
GROUP BY students.major; ''')

df3=pd.DataFrame(c.fetchall(),columns=['major','AvgGpa'])
df3
```

## SELECT the top 2 students with the highest GPA in each major and order the results by major in ascending order, then by GPA in descending order

```python
c.execute(''' SELECT student_id,name,major,gpa,enrollment_date
            FROM (SELECT student_id,name,major,gpa,enrollment_date,
            ROW_NUMBER() OVER (PARTITION BY major ORDER BY gpa DESC) AS rank_within_major
             FROM students)
            WHERE rank_within_major< 3
            ORDER BY major,gpa DESC;''')

df3=pd.DataFrame(c.fetchall(),columns=['student_id','student_name','major','gpa','en_date'])
df3            
```
